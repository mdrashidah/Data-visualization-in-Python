```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

# chapter 2) **Categorical plots**

**Categorical plots are used to plot data using category wise**

/
Now let's discuss using seaborn to plot categorical data!

There are a few main plot types for this:

* boxplot
* violinplot
* stripplot
*  swarmplot
* barplot
* countplot
/
/
-) *Let's go through examples of each!*

In Seaborn, the **palette argument** is used to set the color scheme for your plot.

```python
ds1 = sns.load_dataset("tips")
ds1
```

1. count plot

```python
cp = sns.countplot(y = ds1['smoker'], hue = ds1['sex'])
plt.title("Count plot to check the number of smokers and non-smokers in the dataset")
```

```python
cp = sns.countplot(x = ds1['smoker'], hue = ds1['sex'])
plt.title("Count plot to check the number of smokers and non-smokers in the dataset")
```

2. bar plot

```python
bp1 = sns.barplot(y = ds1['tip'], x = ds1['smoker'], hue= ds1['sex'])
```

* estimators = mean, median, count, sum, etc., they are used to calculate the value of each bar

```python
bp1 = sns.barplot(y = ds1['tip'], x = ds1['smoker'], hue= ds1['sex'], estimator=np.sum)

# mean is the default estimator but we can change it to median, count, sum, etc.

plt.title("Bar plot to check the average tip given by smokers and non-smokers")
plt.xlabel("Smoker/Non-smoker")
plt.ylabel("Average tip")
```

```python
#mean is used to calculate the average tip given by smokers and non-smokers
# median is used to calculate the median tip given by smokers and non-smokers where median is the middle value in a list of numbers
#sum is used to calculate the total tip given by smokers and non-smokers
#count is used to calculate the number of tips given by smokers and non-smokers
```

```python
sns.boxplot(x='total_bill',y='day', data = ds1, palette="rainbow", hue= "sex")
plt.title("Box plot to check the distribution of total bill amount on different days")

# where palette is used to change the color of the box plot
#hue is used to split the data into different categories
```

* use of * **whis** * in boxplot :

In the context of a box plot, "whis" refers to the *whiskers*.

The whiskers are the * *lines that **extend** from the top and bottom of the central box.* *

They are used to show the spread of the data outside of the middle 50%.

```python
sns.boxplot(x='total_bill',y='day', data = ds1, palette="rainbow", hue= "sex", whis = 4.5)
plt.title("Box plot to check the distribution of total bill amount on different days")
```

4_ violin plot

* violin plot is used to show the distribution of the data and its probability density
* it is similar to box plot but it also shows the kernel density estimation of the data

```python
sns.violinplot(y='total_bill',x='day', data = ds1, palette="rainbow", hue= "sex")
plt.title("Violin plot to check the distribution of total bill amount on different days")
```

5. **Swarm plot**

```python
#1st mtd. to write

sns.swarmplot(y= ds1.total_bill, x= ds1.day, palette="rainbow", hue= ds1.sex)
```

```python
#2nd mtd. to write
sns.swarmplot(y='total_bill',x='day',data = ds1, palette="rainbow", hue= "sex")
```

**Overlapping OF the GrapH**

*. . . .(i) with box plot.*

```python
sns.boxplot(x=ds1.total_bill,y = ds1.day, color = 'lightgreen', whis = 4.5)
sns.swarmplot(x=ds1.total_bill,y = ds1.day, hue = ds1.sex)
```

*. . . .(ii) with violin plot.*

```python
sns.violinplot(x=ds1.total_bill,y = ds1.day) # with same color
sns.swarmplot(x=ds1.total_bill,y = ds1.day, hue = ds1.sex)
```

```python
sns.violinplot(x=ds1.total_bill,y = ds1.day,palette= "rainbow") # with different colors
sns.swarmplot(x=ds1.total_bill,y = ds1.day, hue = ds1.sex, linewidth= 1)
# linewidth is used to change the width of the border of the points in the swarm plot
```

```python
sns.violinplot(x=ds1.total_bill,y = ds1.day,color = 'lightgray', scale = "width")
sns.swarmplot(x=ds1.total_bill,y = ds1.day, hue = ds1.sex, size = 4.55, linewidth= 1)
# size is used to change the size of the dots in the swarm plot
```

4. Strip Plot

```python
sns.stripplot(x=ds1.total_bill,y = ds1.day, hue = ds1.sex, linewidth= 1)

# with default jitter
```

where jitter is used to add some random noise to the data points to avoid overlapping of points

```python
sns.stripplot(x=ds1.total_bill,y = ds1.day, hue = ds1.sex, jitter= False)
# without jitter
```

....for changing style of the graph....

```python
sns.stripplot(x=ds1.total_bill,y = ds1.day, hue = ds1.sex, jitter= True, linewidth= 1, size = 5.55, palette= "rainbow", marker = "s", edgecolor = "black")
```

. "with merging type of 2 plots" .

```python
# "with merging type of 2 plots"
sns.violinplot(x=ds1.total_bill,y = ds1.day,palette= "rainbow") # with different colors

sns.stripplot(x=ds1.total_bill,y = ds1.day, hue = ds1.sex, jitter= True, linewidth= 1, size = 5.55, palette= "rainbow", marker = "s", edgecolor = "black")
```
