1. Heat Map


üëÅ üëâ Heatmap in Seaborn = a colored table that shows numbers as colors (darker/lighter = higher/lower values).

So whenever you see it, just remember: ‚ÄúIt‚Äôs like Excel cells, but instead of numbers only, the cells are shaded with colors to make patterns obvious.‚Äù

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
tipscorln = tips[['total_bill','tip','size']]
tipscorln
```

```python
tipscorln.corr() 
# this will show of the correlation matrix
```

```python
sns.heatmap(tipscorln.corr(), annot= True, cmap= 'coolwarm')

# where annot = True will show the values in the boxes
# cmap = 'coolwarm' will show the color of the boxes
```

2. Pivot Table

üëâ **Pivot Table Heatmap** = A pivot table (summary table) with colors added to show patterns, highs, and lows at a glance.

Think: numbers + colors = quick insights.

```python
pivot_flights = flights.pivot_table(index= 'month', columns= 'year', values= 'passengers')
pivot_flights
```

..** **values parameter** **  specifies the column whose data will be used to fill the cells of the new table. It's the numerical data you want to summarize or aggregate. ..

```python
sns.heatmap(pivot_flights)
```

üëáü§ú for customizing the pivot table heat graph

```python
sns.heatmap(pivot_flights, cmap= 'rainbow', linecolor= 'black', linewidths=1)
```

```python
plt.figure(figsize=(15, 10)) # You can adjust the width and height to your liking
sns.heatmap(pivot_flights, cmap= 'rainbow', linecolor= 'black', linewidth= 1, annot = True)
```
