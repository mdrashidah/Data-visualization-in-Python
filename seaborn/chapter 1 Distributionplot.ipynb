# **Distribution plot**

**> We use distribution plots to visualize the distribution of      *quantitative data* .**

*<} Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:*

- distplot/histplot
- jointplot
- pairplot
- rugplot

```python
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
dt = sns.load_dataset("tips")
dt
# this is used to load the dataset
```

```python
dt['size'].unique()


# this is used to find the unique values in the size column
```

1. **distplot/histplot**


```python
sns.displot(dt['total_bill'])
```

```python
sns.histplot(dt['total_bill'])

# we can use hisplot instead of displot also as both are same
```

*/ bins and kde /*

* Bins are contiguous intervals or ranges used to group numerical data. The most common use of bins is in creating a histogram, which is a bar chart that displays the frequency of data points falling into each bin.

* 

```python
dt
```

```python
dt['total_bill'].unique()
```

```python
sns.histplot(dt['total_bill'],bins = 10,kde = True)
# bins is used to increase the number of bars in the histogram 
# kde is used to add the kernel density estimation curve on the histogram
```

-\/     )     where
*  bins is used to increase the number of bars in the histogram
*  kde is used to add the kernel density estimation curve on the histogram

```python
plt.subplot(2,1,1)  # 2 rows 1 column 1st plot
sns.histplot(dt['tip'],bins = 10,kde = True)
plt.title("Tip Distribution")
plt.xlabel("Tip")
plt.ylabel("Count")


plt.subplot(2,1,2)
sns.histplot(dt['total_bill'],bins = 20,kde = True, color= 'green')
plt.xlabel("Total Bill")
plt.ylabel("Count")
```

2.  **joint plot**

```python
sns.jointplot(x= 'total_bill', y= 'tip', data = dt, kind = 'kde', color = 'purple')
```

* kind is used to change the type of plotscatter', 'kde', 'hist', 'hex', 'reg', 'resid',"scatter"
* kde is used to plot the kernel density estimation plot
* hex is used to plot the hexagonal binning plot
* reg is used to plot the regression plot
* resid is used to plot the residual plot
* scatter is used to plot the scatter plot
* hist is used to plot the histogram plot

3. pair plot

pair plot is used to plot the pairwise relationships in a dataset

```python
sns.pairplot(dt)
# pairplot is used to plot the pairwise relationships in a dataset
```

```python
sns.pairplot(dt, hue = 'size')
# hue is used to add the color to the plot based on the category
```

```python
sns.pairplot(dt, hue = 'size',palette='dark')
# palette is used to change the color of the plot based on the category
```
