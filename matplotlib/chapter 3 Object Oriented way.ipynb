# **Object Oriented way**

* In Matplotlib, the Object-Oriented way means you explicitly create and control the different parts of your plot. You start by creating a **Figure** object (like a blank canvas) and then add one or more **Axes** objects (the actual plots) to it. This gives you more flexibility and control, especially for complex figures with multiple plots.

```python
x = np.linspace(0,5,11)
x
y = x **3 
y
```

```python
fig0 = plt.figure() # empty canvas on which you can draw your plots.
```

```python
fig0 = plt.figure()
axis1 = fig0.add_axes([0.1,0.1,0.8,0.8]) # left,bottom,width,height
axis1.plot(x,y)
axis2 = fig0.add_axes([0.2,0.5,0.4,0.3])
axis2.plot(x,y)
```

* Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

* **figsize** is a tuple of the **width** and **height** of the figure in inches dpi is the dots-per-inch (pixel per inch).

* dpi: Stands for "dots per inch" and sets the resolution of the figure. A higher dpi value results in a higher-resolution image.

* these are used for viewing the clear view of the images

```python
fig0 = plt.figure(figsize=(10,8),dpi = 1000)
axis1 = fig0.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig0.add_axes([0.2,0.5,0.4,0.3])
axis2.plot(x,y)
```
