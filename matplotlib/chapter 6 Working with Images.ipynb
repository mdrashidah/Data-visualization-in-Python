```python
img1.shape
```

```python
cropped_img = img1[300:500, 200:400]  # crop: rows 50-200, cols 100-300
plt.imshow(cropped_img)
plt.axis('on')
```

**To choose the right values for cropping, you first need to know the dimensions of the image. Since the image is loaded as a NumPy array, you can get its height and width using the *.shape* attribute.**

* Find the image shape: The img1.shape attribute will return a tuple (height, width, color_channels).

* Crop based on dimensions: Once you know the height and width, you can use slicing img1[start_row:end_row, start_col:end_col] to select the portion you want.
