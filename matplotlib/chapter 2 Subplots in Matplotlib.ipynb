# **Subplots in Matplotlib**

* In **Matplotlib**, subplots are *used* to create multiple plots within a single figure. They allow you to arrange several graphs in a grid of rows and columns, making it easy to compare different data visualizations side-by-side.

```python
x = np.linspace(0, 10, 15)
y1 = x**2
x
y1
```

```python
plt.subplot(2,2,1)  # (rows, columns, panel number)
plt.subplot(2,2,2)
plt.subplot(2,2,3)
plt.subplot(2,2,4)
```

```python
plt.subplot(2,2,1)  # (rows, columns, panel number)
plt.subplot(2,2,2)
plt.plot(x, y1, 'o-')
plt.subplot(2,2,3)
plt.plot(y1,x, 'o-')
plt.subplot(2,2,4)
plt.plot(x, x, 'o-') # this is y = x line a simple line and straight line
```
