```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

## CHAPTER 1) **Matplotlib Basics**

**definition for Matplotlib:**

**Matplotlib** is a popular Python library used for *creating static, animated, and interactive visualizations.* It *provides* a wide variety of plots and charts, like line graphs, bar charts, and scatter plots, to help you understand and present your data visually.

```python
x = np.linspace(0, 10, 15)
x
```

```python
y = x*2
y
```

```python
plt.plot(x, y) # Create a line plot
plt.xlabel('X-axis') # Label for the x-axis
plt.ylabel('Y-axis') # Label for the y-axis
plt.title('Line Plot of y = 2x') # Title of the plot
plt.show() # Display the plot
```

```python
y1 = x**2
y1
```

```python
plt.plot(x, y1, 'o-') # 'o-' specifies circle markers connected by lines
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot of y = x^2')
```
